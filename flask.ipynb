```python
'''
    Q3. What is App routing in Flask? Why do we use app routes?
'''
```

```python
In Flask, app routing refers to the process of mapping a URL to a specific function or view in your application. 
This is accomplished using the @app.route() decorator in your Flask application code.

We use app routes to provide a way for users to interact with our web application by specifying a URL that 
corresponds to a specific function or view in our code. 
This allows users to navigate our web application, perform actions, and retrieve information.
```

```python
'''
    Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
    url_for() function.
'''
```

```python
The url_for() function is used for URL building. It generates a URL to a specific function.
The url_for() function accepts the name of the function as the first argument and any number of
keyword arguments corresponding to the variable parts of the URL.
```

```python
from flask import Flask,url_for
app = Flask(__name__)

@app.route('/')
def home():
    return 'Home page'

@app.route('/about')
def about():
    return 'About page'

with app.test_request_context():
    print(url_for('home'))
    print(url_for('about'))
    
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
